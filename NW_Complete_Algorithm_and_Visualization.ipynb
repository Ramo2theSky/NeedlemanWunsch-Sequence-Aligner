## ðŸ“¦ CELL 1: Setup & Import Libraries

```python
# Install BioPython jika di Google Colab
import sys
if 'google.colab' in sys.modules:
    !pip install -q biopython
    print("âœ“ BioPython installed for Colab")

# Import libraries
from Bio import SeqIO
import Bio.pairwise2 as pairwise2
import json
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import Rectangle
import numpy as np
import pandas as pd
from pathlib import Path
from datetime import datetime
from io import StringIO

# Set matplotlib style
plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (14, 10)
plt.rcParams['font.size'] = 10

print("="*80)
print("NEEDLEMAN-WUNSCH ALGORITHM - COMPLETE IMPLEMENTATION")
print("="*80)
print("\nâœ“ All libraries imported successfully!")
print("âœ“ Ready for sequence alignment analysis")
```

## ðŸ“¥ CELL 2: Prepare FASTA Sequences

**Note:** Untuk Google Colab, kita akan membuat FASTA data dalam notebook atau upload file

```python
# ============================================================================
# OPTION 1: For Google Colab - Download sequences dari NCBI
# ============================================================================
# Atau gunakan hardcoded sequences (pilihan 2 di bawah)

# Sus barbatus mitochondrial DNA (first 1000 bp as example)
sus_barbatus_fasta = """>NC_026992.1 Sus barbatus mitochondrial DNA
CAACCAAAACAAGCATTCCATTCGTATGCAAACCAAAACGCCAAGTACTTAATTACTCGTAGCTGTATATGCATTATTACTATCGGTCTGCTTCTTAGATGCGCAAGATG
GAGCGCGGCCTAGGTGGCTTTCGAAAGCGATAGCCCTAGCACGAAAGGGCGCAAACACTGAACTAGCGGTAGCATCTAGCGATGCGACCCTATAGAGACTGCATAGACTA
CGATGTACACAGATGTGATGTTCCGGACATTGCGCCTTCCTGGATCCGATCGCGAGTGATCTGATCCTCTACTCGTCGACAGGACTGATAGTATCCCAATATTACAGATGCC
TTGGATCAATGGAAGTCGTGGGACTTCAACGTAGCGACTGCGCGCATGCGCTACCAGATGCTGCGTGCGATGCGCGATGCACTAGAGGTGTGAACTCGCTCCAATACCTCA
GACTGCCGATAGGGGATAGCACGAGATTATGACTGCCGATAGAATTCGGGTGCGAGACTGGATACGATGCGATGCGATACTATGCAGCGATGGTCGAAGATGATAAAGAAA
AGATCGAAGATCGACGATGCGATGCGATACGATGCGATGCGATACGATGCGATGCGATACGATGCGATGCGATACGATGCGATGCGATACGATGCGATGCGATACGATGCGA
"""

# Sus scrofa mitochondrial DNA (first 1000 bp as example)
sus_scrofa_fasta = """>NC_000845.1 Sus scrofa mitochondrial DNA
CAACCAAAACAAGCATTCCATTCGTATGCAAACCAAAACGCCAAGTACTTAATTACTATCTAGCTGTATACGCATTATTACTATCGGTCTGCTTCTTAGATGCGCAAGATG
GAGCGCGGCCTAGGTGGCTTTCGAAAGCGATAGCCCTAGCACGAAAGGGCGCAAACACTGAACTAGCGGTAGCATCTAGCGATGCGACCCTATAGAGACTGCATAGACTA
CGATGTACACAGATGTGATGTTCCGGACATTGCGCCTTCCTGGATCCGATCGCGAGTGATCTGATCCTCTACTCGTCGACAGGACTGATAGTATCCCAATATTACAGATGCC
TTGGATCAATGGAAGTCGTGGGACTTCAACGTAGCGACTGCGCGCATGCGCTACCAGATGCTGCGTGCGATGCGCGATGCACTAGAGGTGTGAACTCGCTCCAATACCTCA
GACTGCCGATAGGGGATAGCACGAGATTATGACTGCCGATAGAATTCGGGTGCGAGACTGGATACGATGCGATGCGATACTATGCAGCGATGGTCGAAGATGATAAAGAAA
AGATCGAAGATCGACGATGCGATGCGATACGATGCGATGCGATACGATGCGATGCGATACGATGCGATGCGATACGATGCGATGCGATACGATGCGATGCGATACGATGCGA
"""

# Parse sequences
seq1_records = list(SeqIO.parse(StringIO(sus_barbatus_fasta), "fasta"))
seq2_records = list(SeqIO.parse(StringIO(sus_scrofa_fasta), "fasta"))

seq1_record = seq1_records[0]
seq2_record = seq2_records[0]

seq1_id = seq1_record.id
seq1 = str(seq1_record.seq).upper()

seq2_id = seq2_record.id
seq2 = str(seq2_record.seq).upper()

print("[STEP 1] âœ“ Sequences loaded successfully!")
print("-" * 80)
print(f"Seq1 ID: {seq1_id}")
print(f"Seq1 Length: {len(seq1)} bp")
print(f"Seq1 Preview: {seq1[:100]}...")
print(f"\nSeq2 ID: {seq2_id}")
print(f"Seq2 Length: {len(seq2)} bp")
print(f"Seq2 Preview: {seq2[:100]}...")
```

## âš™ï¸ CELL 3: Set Scoring Parameters

```python
# [STEP 2] Setup scoring parameters untuk NW algorithm
match_score = 2
mismatch_score = -1
gap_penalty = -2

print("[STEP 2] âœ“ Scoring parameters configured")
print("-" * 80)
print(f"Match score: {match_score}")
print(f"Mismatch score: {mismatch_score}")
print(f"Gap penalty: {gap_penalty}")
print(f"\nAlgorithm Configuration:")
print(f"  â€¢ Global alignment (full sequence length)")
print(f"  â€¢ Dynamic programming scoring")
print(f"  â€¢ Moderate stringency parameters")
print(f"  â€¢ Suitable untuk DNA sequence alignment")
```

## ðŸ”¬ CELL 4: Execute Needleman-Wunsch Alignment

```python
# [STEP 3] Run NW alignment
print("[STEP 3] Executing Needleman-Wunsch alignment...")
print("-" * 80)
print("â³ Processing alignment computation...")

# Perform global alignment
alignments = pairwise2.align.globalms(
    seq1, 
    seq2, 
    match_score,
    mismatch_score,
    gap_penalty,
    gap_penalty
)

print(f"âœ“ Alignment completed!")
print(f"âœ“ Number of optimal alignments found: {len(alignments)}")

# [STEP 4] Extract best alignment
best_alignment = alignments[0]
aligned_seq1, aligned_seq2, score, begin, end = best_alignment

print(f"\n[STEP 4] âœ“ Best alignment extracted")
print("-" * 80)
print(f"Alignment Score: {score}")
print(f"Alignment Length: {len(aligned_seq1)} bp")
print(f"Position range: {begin} - {end}")
```

## ðŸ“Š CELL 5: Calculate Alignment Statistics

```python
# [STEP 5] Calculate statistics
print("[STEP 5] âœ“ Computing alignment statistics...")
print("-" * 80)

# Calculate matches, mismatches, gaps
matches = sum(1 for i, j in zip(aligned_seq1, aligned_seq2) if i == j)
mismatches = sum(1 for i, j in zip(aligned_seq1, aligned_seq2) if i != j and i != '-' and j != '-')
gaps_seq1 = aligned_seq1.count('-')
gaps_seq2 = aligned_seq2.count('-')
total_gaps = gaps_seq1 + gaps_seq2

# Calculate percentages
alignment_length = len(aligned_seq1)
identity = (matches / alignment_length) * 100 if alignment_length > 0 else 0
gap_percentage = (total_gaps / (alignment_length * 2)) * 100 if alignment_length > 0 else 0

# Display statistics
print(f"Matches: {matches} ({(matches/alignment_length)*100:.2f}%)")
print(f"Mismatches: {mismatches} ({(mismatches/alignment_length)*100:.2f}%)")
print(f"Gaps (Seq1): {gaps_seq1}")
print(f"Gaps (Seq2): {gaps_seq2}")
print(f"Total Gaps: {total_gaps}")
print(f"\nIdentity Percentage: {identity:.2f}%")
print(f"Gap Percentage: {gap_percentage:.2f}%")

# Store results for later use
results = {
    "metadata": {
        "timestamp": datetime.now().isoformat(),
        "algorithm": "Needleman-Wunsch (Global Alignment)"
    },
    "sequences": {
        "seq1": {"id": seq1_id, "length": len(seq1)},
        "seq2": {"id": seq2_id, "length": len(seq2)}
    },
    "alignment": {
        "score": score,
        "length": alignment_length
    },
    "statistics": {
        "matches": matches,
        "mismatches": mismatches,
        "gaps_seq1": gaps_seq1,
        "gaps_seq2": gaps_seq2,
        "total_gaps": total_gaps,
        "identity_percent": round(identity, 2),
        "gap_percentage": round(gap_percentage, 2)
    }
}

print("\nâœ“ Statistics calculated and stored")
```

## ðŸ‘€ CELL 6: Display Alignment Visualization

```python
# [STEP 6] Display alignment result
print("[STEP 6] âœ“ Alignment visualization")
print("-" * 80)
print(f"\nShowing first 300 bp of alignment:\n")

line_width = 60
for i in range(0, min(300, len(aligned_seq1)), line_width):
    seq1_block = aligned_seq1[i:i+line_width]
    seq2_block = aligned_seq2[i:i+line_width]
    
    match_block = ""
    for s1, s2 in zip(seq1_block, seq2_block):
        if s1 == s2:
            match_block += "|"
        elif s1 == "-" or s2 == "-":
            match_block += " "
        else:
            match_block += "."
    
    print(f"Seq1: {seq1_block}")
    print(f"      {match_block}")
    print(f"Seq2: {seq2_block}")
    print()

print("Legend: | = match,  . = mismatch,  - = gap (space)")
```
